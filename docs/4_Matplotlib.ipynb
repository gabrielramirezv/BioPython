```python
# Importar librerias
import pandas as pd

# Leer los datos desde un archivo CSV
data = pd.read_csv("../data/gapminder_gdp_americas.csv", index_col = "country")
# Eliminar columnas no necesarias
data_america = data.iloc[:,1:]
# Adaptar el nombre de las columnas
year = data_america.columns.str.replace('gdpPercap_', '')
data_america.columns = year.astype(int)


plt.clf()
plt.style.use('ggplot')
data_america.T.plot(kind='bar')
plt.ylabel("GDP per capita")
plt.show()
```

```python
import pandas as pd

gdp_mex = data_america.loc['Mexico']
plt.clf()
plt.plot(year, gdp_mex, 'r--')
plt.show()
```

```python
# Importar librerias
import pandas as pd

gdp_mex = data_america.loc['Mexico']
gdp_pan = data_america.loc['Panama']

plt.clf()
plt.plot(year, gdp_mex, 'r-', label="Mexico")
plt.plot(year, gdp_pan, 'b-', label="Panama")

plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')

plt.show()
```
